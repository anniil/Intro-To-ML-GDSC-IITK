## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python

import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.arange(0, 100)
y = x * 2

# Exercise 1
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='blue', lw=2, ls='--')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Plot of x versus 2x')

plt.show()

```

**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.arange(0, 100)
y = x * 2

# Exercise 2
fig = plt.figure()

# Main axis (ax1)
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, y, color='blue')
ax1.set_xlabel('X-axis (Main)')
ax1.set_ylabel('Y-axis (Main)')

# Inset axis (ax2)
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax2.plot(x, y, color='orange')
ax2.set_xlabel('X-axis (Inset)')
ax2.set_ylabel('Y-axis (Inset)')

plt.show()

```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.arange(0, 100)
y = x * 2
z = x ** 2

# Exercise 4
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Plot on the first subplot (axes[0])
axes[0].plot(x, y, color='blue', linewidth=2, linestyle='--')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].set_title('Plot 1')

# Plot on the second subplot (axes[1])
axes[1].plot(x, z, color='red', linewidth=3)
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('Z-axis')
axes[1].set_title('Plot 2')

# Display the figure
plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.arange(0, 100)
y = x * 2
z = x ** 2

# Exercise 4
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Plot (x, y) on the first subplot (axes[0])
axes[0].plot(x, y, color='blue', linewidth=2, linestyle='--', label='Line 1')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].set_title('Plot 1')
axes[0].legend()

# Plot (x, z) on the second subplot (axes[1])
axes[1].plot(x, z, color='red', linewidth=3, linestyle=':', label='Line 2')
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('Z-axis')
axes[1].set_title('Plot 2')
axes[1].legend()

# Display the figure
plt.show()

```
