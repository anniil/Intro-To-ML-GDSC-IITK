## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure() 
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, 'b')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("title")
```

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig = plt.figure()
fig.tight_layout()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.set_xticks([.0, .2, .4, .6, .8, 1.0])
ax1.set_yticks([.0, .2, .4, .6, .8, 1.0])
ax2.set_xticks([.0, .2, .4, .6, .8, 1.0])
ax2.set_yticks([.0, .2, .4, .6, .8, 1.0])
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
ax1.plot(x, y, label= 'fg')
ax2.plot(x, y, label = 'ef')
plt.show()
fig
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # main axes
axes2 = fig.add_axes([.2, .5, .4, .4]) # inset axes

```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
k = np.arange(0,100,1)
l = k**2
# Larger Figure Axes 1
axes1.plot(k, l, 'b')
axes1.set_xlabel('X')
axes1.set_xlim(0, 100)
axes1.set_ylim(0, 10000)
axes1.set_ylabel('Z')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(k, k*2, 'b')
axes2.set_xlim(20.0, 22.0)
axes2.set_ylim(30, 50)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Axes 2 Title')
fig
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
%matplotlib inline
fig4, axis = plt.subplots(1,2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
k = np.arange(0,100,1)
l = k**2
axis[0].plot([0,100], [0,200], 'b', lw =3, ls = "--")
axis[1].plot(k, l, 'g', lw =1, ls = "-")
fig4
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
%matplotlib inline
fig4, axis = plt.subplots(1,2, figsize = (10, 2))
k = np.arange(0,100,1)
l = k**2
axis[0].plot([0,100], [0,200], 'b', lw =3, ls = "-")
axis[1].plot(k, l, 'g', lw =2, ls = "--")
```
