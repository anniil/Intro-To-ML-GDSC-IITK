```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a joint plot with scatter plot and histograms
sns.jointplot(x='fare', y='age', data=titanic)

# Set labels and title
plt.xlabel('Fare')
plt.ylabel('Age')
plt.suptitle('Joint Plot: Fare vs Age', y=1.02)  # Adjust the title position

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create bins with a gap of 20 in 'Fare' values
fare_bins = pd.cut(titanic['fare'], bins=range(0, int(titanic['fare'].max()) + 21, 20), right=False)

# Create a bar plot for the distribution of 'Fare' values with a gap of 20
sns.countplot(x=fare_bins, data=titanic, palette='viridis')

# Set labels and title
plt.xlabel('Fare Bins (with a gap of 20)')
plt.ylabel('Count of Values')
plt.title('Distribution of Fare Values with a Gap of 20')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a box plot for Age vs Class
sns.boxplot(x='class', y='age', data=titanic, palette='viridis')

# Set labels and title
plt.xlabel('Class')
plt.ylabel('Age')
plt.title('Box Plot: Age vs Class')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a swarm plot for Age vs Class
sns.swarmplot(x='class', y='age', data=titanic, palette='viridis')

# Set labels and title
plt.xlabel('Class')
plt.ylabel('Age')
plt.title('Swarm Plot: Age vs Class')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a bar plot for Count vs Sex
sns.countplot(x='sex', data=titanic, palette='viridis')

# Set labels and title
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Count of Passengers by Sex')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Filter numerical columns for correlation matrix
numerical_columns = titanic.select_dtypes(include=['float64', 'int64'])

# Create a correlation matrix
correlation_matrix = numerical_columns.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

# Set title
plt.title('Correlation Heatmap of Titanic Columns')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the distribution of Age for males
sns.histplot(titanic[titanic['sex'] == 'male']['age'].dropna(), ax=axes[0], color='blue')
axes[0].set_title('Age Distribution - Males')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

# Plot the distribution of Age for females
sns.histplot(titanic[titanic['sex'] == 'female']['age'].dropna(),  ax=axes[1], color='purple')
axes[1].set_title('Age Distribution - Females')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
