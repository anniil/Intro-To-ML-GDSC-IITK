```python
# Create a JointGrid with a bar plot and a strip plot
g = sns.jointplot( y="age", x="fare",data=titanic)
g.ax_joint.set_yticks([0, 20, 40, 60, 80, 100])
g.ax_joint.set_xticks([-100,0,100, 200, 300, 400, 500, 600])

# # Add a bar plot on the top axis
# #g = g.plot(sns.barplot, color="b", edgecolor="w")

# # Add a strip plot on the right axis
# #g = g.plot(sns.stripplot, color="b", edgecolor="w")
# #g = g.plot(sns.regplot, sns.distplot)
# # Show the plot
plt.show()
```

```python
b=sns.distplot(titanic['fare'],kde=False,bins=30,color='r')
b.set_xlim([0,600])
```

```python
sns.boxplot(x="class", y="age", data=titanic,palette='rainbow')
```

```python
custom_colors = ["mediumaquamarine", "coral", "lightsteelblue"]
sns.swarmplot(x="class", y="age", data=titanic,palette=custom_colors,alpha=0.9)
```

```python
colour=["cornflowerblue","mediumseagreen"]
sns.countplot(x='sex',data=titanic,palette=colour)
```

```python
a=sns.heatmap(titanic.corr(),cmap='coolwarm',cbar_kws={'ticks': [-0.8,-0.4,0.0,0.4,0.8]},vmax=0.8,vmin=-0.8)
a.set_title('titanic.corr()')
```

```python
t=sns.FacetGrid(titanic,col="sex")
t=t.map(plt.hist,"age",color='steelblue')
t.set(xticks=range(0,81,10),xlim=[0,80])
t.fig.subplots_adjust(wspace=0.1)
```
