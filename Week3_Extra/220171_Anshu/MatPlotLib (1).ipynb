```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.1, 0.6, 0.4, 0.2]) # inset axes
axes2.set_xticks([20, 20.5, 21, 21.5, 22, 22.5])  # Set the tick positions on the x-axis
axes2.set_yticks([30, 35, 40, 45, 50])  # Set the tick positions on the y-axis

# Larger Figure Axes 1
axes1.plot(x, z, 'b')
axes1.set_xlabel('X')
axes1.set_ylabel('Z')
#axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(x, y, 'b')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoom');
axes2.set_xlim(20.0,22.0)
axes2.set_ylim(30,50)
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(5, 3))

# Plot on the first subplot
axes[0].plot(x, x*2, 'b--',label='Linear', linewidth=1.00)

axes[0].set_xticks([0,50,100,75,25 ])
axes[1].set_xticks([0,50,100,75,25 ])

axes[0].set_yticks([0,25,50,75,100,125,150,175,200])
axes[1].set_yticks([0,2500,5000,7500,10000 ])
# Plot on the second subplot
axes[1].plot(x, x**2,'r', label='Quadratic',linewidth=1.00)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```
