## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel(' x ') # Notice the use of set_ to begin methods
axes.set_ylabel('y')
axes.set_title('title')
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python


# Create Figure (empty canvas)
fig = plt.figure()

# Add the first set of axes to figure [0,0,1,1]
axes1 = fig.add_axes([0, 0, 1, 1])  # Main axes

# Add the second set of axes to figure [0.2,0.5,.4,.4]
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])  # Inset axes

# Plot on the main set of axes (axes1)
axes1.plot(x, y, 'b')
axes1.set_xlabel('X Label for Main Plot')
axes1.set_ylabel('Y Label for Main Plot')
axes1.set_title('Main Title')

# Plot on the inset set of axes (axes2)
axes2.plot(y, x, 'r')
axes2.set_xlabel('X Label for Inset Plot')
axes2.set_ylabel('Y Label for Inset Plot')
axes2.set_title('Inset Title')

plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt

# Create Figure (empty canvas)
fig, ax1 = plt.subplots()

# Plot on the main set of axes (ax1)
ax1.plot(x, z, 'b')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

# Create inset of the larger plot (ax2)
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])  # [left, bottom, width, height]

# Plot on the inset set of axes (ax2)
ax2.plot(x, y, 'r')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoomed In')

# Set specific x and y limits for the inset plot
ax2.set_xlim(20, 22)  # Example x limits
ax2.set_ylim(30, 50)  # Example y limits

plt.show()
```
